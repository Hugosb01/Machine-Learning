## 6) Utilidad para métricas (MSE, MAE, RMSE, MAPE, R²)


```python

def regression_metrics(y_true, y_pred):
    """  
    Calcula las métricas de regresión: MAE, MSE, RMSE, MAPE, R²  
    
    Parameters:
    y_true: valores reales
    y_pred: valores predichos por el modelo
    
    Returns:
    dict: Diccionario con todas las métricas
    """ 
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = root_mean_squared_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    return {"MAE": mae, "MSE": mse, "RMSE": rmse, "MAPE": mape, "R2": r2}
```

### MAS PREGUNTAS PARA RESPONDER:
1. Que aporta cada metrica?
    Mae --> Mide el error promedio en términos absolutos
    Mse --> Mide el promedio de los errores al cuadrado
    Rmse --> Es la raíz cuadrada del MSE
    Mape --> Mide el error en porcentaje
    R2 --> Mide qué porcentaje de la variabilidad explica el modelo
2. Cual es la mas importante?
En realidad, todos son importantes y dependerá del tipo de datos con los que estemos trabajando y que conclusiones queramos sacar. 
En este caso, podríamos hablar de que quizás el R^2 para saber que tan bien explica el peso y el MAPE para saber en que porcentaje promedio se equivoca.

## 7) KNNRegressor

https://www.ibm.com/es-es/think/topics/knn

Instancia proporcionada. **Completa**: GridSearchCV (parrilla típica), `fit`, `predict`, métricas y guarda en `results` para comparar al final.


```python

# =====================
# Instanciación del modelo
# =====================
#Explicación del modelo
"""KNN predice el peso de un pez nuevo encontrando los "k" peces más parecidos en el dataset y promediando sus pesos. 
la "cercanía" se mide con distancias matemáticas (euclídea, manhattan) y puede ponderar más a los vecinos más próximos."""
# Creamos una variable con el modelo de KNNRegressor 
model = KNeighborsRegressor()

# =====================
# TODO: GridSearchCV
# =====================
#Creamos un diccionario con los parametros a probar en el GridSearchCV
param_grid = {
    #Probamos con estos parametros
    'n_neighbors': [3, 5, 7, 9], # Numero de vecinos
    'weights': ['uniform', 'distance'],# Peso uniforme o por distancia
    'metric': ['euclidean', 'manhattan']# Tipo de distancia
    }  
# GrindSearchCV es un entrenador inteligente que prueba todas las combinaciones de parametros y selecciona la mejor
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
#estimator=model es el modelo que queremos entrenar, en este caso KNNRegressor
#param_grid=param_grid es el diccionario con los parametros a probar
#cv=5 es el numero de particiones. Usa 5 particiones para validar el modelo, 4 de ellas para entrenar y 1 para validar, y repite el proceso 5 veces
#n_jobs=-1 usa todos los nucleos del procesador para entrenar el modelo
#scoring='neg_mean_squared_error' es la metrica que queremos optimizar, en este caso el error cuadratico medio negativo (negativo porque GridSearchCV maximiza la metrica, y nosotros queremos minimizar el error)

grid_search.fit(x_train_scaled, y_train)
best_model = grid_search.best_estimator_ #Utilizamos la función best_estimator_ para obtener el mejor modelo entrenado
print("Best parameters:", grid_search.best_params_) 



# =====================
# TODO: Predicción y métricas
# =====================
y_pred = best_model.predict(x_test)
mets = regression_metrics(y_test, y_pred)
print("Metrics:", mets)     

results.append({"modelo": "KNNRegressor", **mets}) # guardamos los resultados de las metricas por si queremos comparar
```

## 7) SVR

https://www.ibm.com/es-es/think/topics/support-vector-machine

Instancia proporcionada. **Completa**: GridSearchCV (parrilla típica), `fit`, `predict`, métricas y guarda en `results` para comparar al final.


```python

# =====================
# Instanciación del modelo
# =====================
#Explicación del modelo
"""SVR encuentra la línea (o superficie) que mejor se ajusta a los datos, pero permitiendo un margen de error (épsilon) donde no penaliza desviaciones pequeñas, 
y usa "kernels" para capturar relaciones  entre las dimensiones de los peces y su peso"""
model = SVR()

# =====================
# TODO: GridSearchCV
# =====================
param_grid = {
    #Probamos con estos parametros
    'kernel': ['linear', 'poly', 'rbf'], # Tipo de kernel
    'C': [0.1, 1, 10, 100], # Regularización
    'epsilon': [0.1, 0.2, 0.5, 1], # Margen de tolerancia
    'gamma': ['scale', 'auto'] # Coeficiente del kernel
    }
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(x_train_scaled, y_train)
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)     



# =====================
# TODO: Predicción y métricas
# =====================
y_pred = best_model.predict(x_test)
mets = regression_metrics(y_test, y_pred)
print("Metrics:", mets)     
results.append({"modelo": "SVR", **mets}) # guardamos los resultados de las metricas por si queremos comparar   
```

### PREGUNTAS PARA RESPONDER:
1. ¿Qué papel juega el parámetro `C` en SVR?
Sirve para regularizar los datos, en función de si el rango de c es pequeño o grande, permetirá más o menos precisión

2. ¿Y el parámetro `epsilon`?
Actúa como una especie de banda donde entran los errores. Cuanto más pequeña sea, más pequeños más precisos deben de los resultados para considerarse
y viceversa 

3. ¿Qué diferencias hay entre los kernels `linear`, `poly` y `rbf`?
Básicamente cada uno encuentra relaciones de diferente naturaleza: lineal, polinómica o compleja 

## 7) DecisionTreeRegressor

https://www.ibm.com/es-es/think/topics/decision-trees

Instancia proporcionada. **Completa**: GridSearchCV (parrilla típica), `fit`, `predict`, métricas y guarda en `results` para comparar al final.

```python

# =====================
# Instanciación del modelo
# =====================
#Explicación del modelo
""" El árbol de decisiones predice el peso de los peces haciendo una serie de preguntas binarias sobre sus dimensiones (¿Length1 > 25?, ¿Height < 12?) hasta llegar a una hoja
 que contiene el peso promedio de peces similares, creando una estructura de reglas fácil de interpretar."""
model = DecisionTreeRegressor(random_state=42)

# =====================
# TODO: GridSearchCV
# =====================

param_grid = {
    'max_depth': [None, 5, 10, 20], # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10], # Mínimo número de muestras para dividir un nodo
    'min_samples_leaf': [1, 2, 4] # Mínimo número de muestras en una hoja
    }
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(x_train_scaled, y_train)
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)                     


# =====================
# TODO: Predicción y métricas
# =====================
y_pred = best_model.predict(x_test)
mets = regression_metrics(y_test, y_pred)
print("Metrics:", mets)     
results.append({"modelo": "DecisionTreeRegressor", **mets}) # guardamos los resultados de las metricas por si queremos comparar 
```

### PREGUNTAS PARA RESPONDER:
1.- El profe dice que los árboles no necesitan escalado, ¿por qué?
Básicamente por que funcionan por comparaciones de umbrales y por lo tanto simplemente si hay grandes diferencias, lo va a saber diferenciar 
2.- ¿Qué papel juega el parámetro `max_depth`?
Controla la profundidad máxima del árbol, es decir, cuántas preguntas consecutivas puede hacer antes llegar a una respuesta

## 7) RandomForestRegressor

https://www.ibm.com/es-es/think/topics/random-forest

Instancia proporcionada. **Completa**: GridSearchCV (parrilla típica), `fit`, `predict`, métricas y guarda en `results` para comparar al final.

```python

# =====================
# Instanciación del modelo
# =====================
#Explicación del modelo
""" Random Forest crea múltiples árboles de decisión (un "bosque") donde cada árbol ve una versión ligeramente diferente de los datos y features, 
luego combina sus predicciones promediándolas para obtener una estimación más robusta y estable que un solo árbol."""
model = RandomForestRegressor(random_state=42)

# =====================
# TODO: GridSearchCV
# =====================
param_grid = {
    'n_estimators': [50, 100, 200], # Número de árboles en el bosque
    'max_depth': [None, 5, 10, 20], # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10], # Mínimo número de muestras para dividir un nodo
    'min_samples_leaf': [1, 2, 4] # Mínimo número de muestras en una hoja
    }
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(x_train_scaled, y_train)
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)     


# =====================
# TODO: Predicción y métricas
# =====================

y_pred = best_model.predict(x_test)
mets = regression_metrics(y_test, y_pred)
print("Metrics:", mets)     
results.append({"modelo": "RandomForestRegressor", **mets}) # guardamos los resultados de las metricas por si queremos comparar
```
